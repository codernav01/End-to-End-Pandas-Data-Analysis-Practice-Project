###  Section D: Outlier Detection & Treatment (36–50)

* Identified and treated outliers using statistical methods such as IQR.
* This enhances data consistency and minimizes the impact of anomalies.

```python
# Detect outliers in age using the IQR method.

Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['age'] < lower) | (df['age'] > upper)]
outliers
```

```python
# Detect outliers in salary using the IQR method.

Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['salary'] < lower) | (df['salary'] > upper)]
outliers
```

```python
# Identify rows where projects is an outlier.

Q1 = df['projects'].quantile(0.25)
Q3 = df['projects'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df[(df['projects'] < lower) | (df['projects'] > upper)]
```

```python
# Create a new column flagging salary outliers.

Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['salary_outlier'] = ((df['salary'] < lower) | (df['salary'] > upper))
df['salary_outlier']
```

```python
# Remove rows containing salary outliers.

df = df[~df['salary_outlier']]
df
```

```python
# Cap salary outliers using upper and lower IQR limits.

Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['salary'] = df['salary'].clip(lower, upper)
df['salary']
```

```python
# Replace age outliers with median age.

Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

median_age = df['age'].median()

df.loc[(df['age'] < lower) | (df['age'] > upper), 'age'] = median_age
```

```python
# Detect outliers in rating.

Q1 = df['rating'].quantile(0.25)
Q3 = df['rating'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df[(df['rating'] < lower) | (df['rating'] > upper)]
```

```python
# Detect department-wise salary outliers.

def detect_outliers(group):
    Q1 = group['salary'].quantile(0.25)
    Q3 = group['salary'].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return group[(group['salary'] < lower) | (group['salary'] > upper)]

df.groupby('department', observed=True, group_keys=False)\
  .apply(detect_outliers, include_groups=False)
```

```python
# Count number of outliers in each numeric column.

num_cols = df.select_dtypes(include='number').columns

outlier_counts = {}

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outlier_counts[col] = ((df[col] < lower) | (df[col] > upper)).sum()

outlier_counts
```

```python
# Identify rows having more than one outlier.

num_cols = df.select_dtypes(include='number').columns

outlier_flag = pd.DataFrame()

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outlier_flag[col] = (df[col] < lower) | (df[col] > upper)

# Rows with more than 1 outlier
df[outlier_flag.sum(axis=1) > 1]
```

```python
# Replace outliers in projects with the 95th percentile.

Q1 = df['projects'].quantile(0.25)
Q3 = df['projects'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

p95 = df['projects'].quantile(0.95)

df.loc[(df['projects'] < lower) | (df['projects'] > upper), 'projects'] = p95
```

```python
#Normalize salary after outlier treatment.

df['salary'] = (df['salary'] - df['salary'].min()) / (df['salary'].max() - df['salary'].min())
df['salary']
```

```python
# Compare salary before and after capping.

# Step 1: Original salary save karo
df['salary_before'] = df['salary']

# Step 2: Capping karo
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['salary'] = df['salary'].clip(lower, upper)

# Step 3: Compare
df[['salary_before', 'salary']]
```

```python
#Create a clean DataFrame with no outliers.

num_cols = df.select_dtypes(include='number').columns

clean_df = df.copy()

for col in num_cols:
    Q1 = clean_df[col].quantile(0.25)
    Q3 = clean_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    clean_df = clean_df[(clean_df[col] >= lower) & (clean_df[col] <= upper)]

clean_df
```
